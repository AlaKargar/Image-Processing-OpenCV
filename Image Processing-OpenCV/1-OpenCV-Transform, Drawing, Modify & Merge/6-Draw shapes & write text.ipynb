```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Draw shapes

```python
square= np.zeros((300, 300), np.uint8)
cv2.rectangle(square, (50, 50), (250, 250), 255, -1)

ellipse= np.zeros((300, 300), np.uint8)
cv2.ellipse(ellipse, (150, 150), (150, 150), 30, 0, 180, 255, -1)

plt.subplot(121);plt.imshow(square, cmap= 'gray');plt.title('Square')
plt.subplot(122);plt.imshow(ellipse, cmap= 'gray');plt.title('Ellipse')
```

Bitwise Opration

```python
bitwise_and= cv2.bitwise_and(square, ellipse)
bitwise_or= cv2.bitwise_or(square, ellipse)
bitwise_xor= cv2.bitwise_xor(square, ellipse)
bitwise_not= cv2.bitwise_not(square)

plt.Figure(figsize=[12, 6])
plt.subplot(242);plt.imshow(square, cmap='gray');plt.title('Square')
plt.subplot(243);plt.imshow(ellipse, cmap='gray');plt.title('Ellipse')
plt.subplot(245);plt.imshow(bitwise_and, cmap='gray');plt.title('And')
plt.subplot(246);plt.imshow(bitwise_or, cmap='gray');plt.title('Or')
plt.subplot(247);plt.imshow(bitwise_xor, cmap='gray');plt.title('Xor')
plt.subplot(248);plt.imshow(bitwise_not, cmap='gray');plt.title('Not (Square)')
```

Crop image with shape

```python
img4= cv2.imread('../Images/4.jpg')
cropped= img4[100:600, 150:650]

circle4= np.zeros((500, 500, 3), np.uint8)
cv2.circle(circle4, (250, 250), 250, (255, 255, 255), -1)

output_img4= cv2.bitwise_and(cropped, circle4)
plt.imshow(output_img4[..., ::-1])
```

```python
png_img4= np.zeros((500, 500, 4), np.uint8)
png_img4[:, :, 0:3]= cropped

circle4_2= np.zeros((500, 500), np.uint8)
cv2.circle(circle4_2, (250, 250), 250, (255, 255, 255), -1)

png_img4[:, :, 3]= circle4_2
cv2.imwrite("Circle.png", png_img4)
```

Draw polygon

```python
img7= np.zeros((512, 512, 3), np.uint8)
pts= np.array([[10, 50], [400, 50], [90, 200], [50, 500]], np.int32)
pts= pts.reshape((-1, 1, 2))

cv2.polylines(img7, [pts], True, (0, 0, 255), 4)
cv2.imshow('Polygon', img7)
plt.imshow(img7[..., ::-1])
cv2.waitKey()
cv2.destroyAllWindows()
```

Write persian text

```python
import arabic_reshaper
from bidi.algorithm import get_display
from PIL import Image, ImageFont, ImageDraw

img9= np.zeros((512, 512, 3), np.uint8)

text_p=  'سلام دنیا'
reshaped_text= arabic_reshaper.reshape(text_p)
bidi_text= get_display(reshaped_text)

text= bidi_text.encode().decode('utf-8')

pil_image= Image.fromarray(img9)
font= ImageFont.truetype('arial', 100, encoding='unic')
draw= ImageDraw.Draw(pil_image)
draw.text((120, 200), text, (150, 0, 200), font=font)

plt.imshow(pil_image)
```
