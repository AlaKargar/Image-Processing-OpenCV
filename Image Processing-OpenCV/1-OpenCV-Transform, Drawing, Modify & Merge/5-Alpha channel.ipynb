```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Show alpha Channel

```python
img6= cv2.imread('../Images/6.png', cv2.IMREAD_UNCHANGED)
img6.shape
```

```python
bgr_img6= img6[..., 0:3]
alpha_img6= img6[..., -1]
rgb_img6= cv2.cvtColor(bgr_img6, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img6); plt.title("RGB");
```
