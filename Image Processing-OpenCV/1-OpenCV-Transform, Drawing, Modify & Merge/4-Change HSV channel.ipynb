```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Change BGR to HSV

```python
img3= cv2.imread('../Images/3.jpg')
hsv_img3= cv2.cvtColor(img3, cv2.COLOR_BGR2HSV)

cv2.namedWindow('HSV image', cv2.WINDOW_NORMAL)
cv2.namedWindow('Hue', cv2.WINDOW_NORMAL)
cv2.namedWindow('Saturation', cv2.WINDOW_NORMAL)
cv2.namedWindow('Value', cv2.WINDOW_NORMAL)

cv2.imshow('HSV image', hsv_img3)
cv2.imshow('Hue', hsv_img3[:, :, 0])
cv2.imshow('Saturation', hsv_img3[:, :, 1])
cv2.imshow('Value', hsv_img3[:, :, 2])

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.subplots_adjust(wspace= 0.5)
plt.subplot(131);plt.imshow(hsv_img3[:, :, 0], cmap= 'gray');plt.title('Hue');
plt.subplot(132);plt.imshow(hsv_img3[:, :, 1], cmap= 'gray');plt.title('Saturation');
plt.subplot(133);plt.imshow(hsv_img3[:, :, 2], cmap= 'gray');plt.title('Value');
```

Saturated image

```python
saturated_img3= cv2.merge([hsv_img3[:, :, 0], cv2.add(hsv_img3[:, :, 1], 50), hsv_img3[:, :, 2]])
saturated_img3= cv2.cvtColor(saturated_img3, cv2.COLOR_HSV2BGR)

plt.Figure(figsize= [12, 5])

plt.subplot(121);plt.imshow(saturated_img3[..., ::-1]);plt.title('Saturated');
plt.subplot(122);plt.imshow(img3[..., ::-1]);plt.title('Orginal');
```

Brighter image

```python
brighter_img3= cv2.merge([hsv_img3[:, :, 0], hsv_img3[:, :, 1], cv2.add(hsv_img3[:, :, 2], 50)])
brighter_img3= cv2.cvtColor(brighter_img3, cv2.COLOR_HSV2BGR)

plt.Figure(figsize= [12, 5])

plt.subplot(121);plt.imshow(brighter_img3[..., ::-1]);plt.title('Bright');
plt.subplot(122);plt.imshow(img3[..., ::-1]);plt.title('Orginal');
```

Change Hue

```python
Hue_img3= cv2.merge([cv2.add(hsv_img3[:, :, 0], 50), hsv_img3[:, :, 1], hsv_img3[:, :, 2]])
Hue_img3= cv2.cvtColor(Hue_img3, cv2.COLOR_HSV2BGR)

plt.Figure(figsize= [12, 5])

plt.subplot(121);plt.imshow(Hue_img3[..., ::-1]);plt.title('Hue');
plt.subplot(122);plt.imshow(img3[..., ::-1]);plt.title('Orginal');
```
