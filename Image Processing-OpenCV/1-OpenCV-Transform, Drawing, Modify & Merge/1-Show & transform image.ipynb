```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Open & Show Image

```python
img= cv2.imread('../Images/1.jpg')
# cv2.namedWindow('It is Radan', cv2.WINDOW_NORMAL)
cv2.imshow('It is Radan', img)
cv2.waitKey()
cv2.destroyAllWindows()
```

Image Type

```python
type(img)
```

Image Shape

```python
print(img.shape)
```

Height & Width of image

```python
print('Height of image:', int(img.shape[0]), 'Pixels')
print('Width of image:', int(img.shape[1]), 'Pixels')
```

Show the image

```python
plt.imshow(img)
```

Change BGR to RGB (method 1)

```python
plt.imshow(img[..., ::-1])
```

Change BGR to RGB (method 2)

```python
img_rgb= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

Change pixels

```python
img[1500:1750,1500:1750] = (150, 0, 255)
plt.imshow(img[..., ::-1])
```

Rotate image

```python
rotate_img= cv2.rotate(img, cv2.ROTATE_180)
cv2.namedWindow('rotate', cv2.WINDOW_NORMAL)
cv2.imshow('rotate', rotate_img)
cv2.waitKey()
cv2.destroyAllWindows()

cv2.imwrite('1-2.png', rotate_img)
```

Flip image

```python
flipped_img_h= cv2.flip(img, 1)
flipped_img_v= cv2.flip(img, 0)
cv2.namedWindow('Flip1', cv2.WINDOW_NORMAL)
cv2.namedWindow('Flip2', cv2.WINDOW_NORMAL)
cv2.imshow('Flip1', flipped_img_h)
cv2.imshow('Flip2', flipped_img_v)
cv2.waitKey()
cv2.destroyAllWindows()

cv2.imwrite('1-3.png', flipped_img_h)
cv2.imwrite('1-4.png', flipped_img_v)
```

Resize & Blur image

```python
resize_img= cv2.resize(img, (500, 500))
blur_img= cv2.blur(resize_img, (2, 5))
#cv2.namedWindow('new', cv2.WINDOW_NORMAL)
cv2.imshow('new', blur_img)
cv2.waitKey()
cv2.destroyAllWindows()

cv2.imwrite('1-5.png', blur_img)
```
