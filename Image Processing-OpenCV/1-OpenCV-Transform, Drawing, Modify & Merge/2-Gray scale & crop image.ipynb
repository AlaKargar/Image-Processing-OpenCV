```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Import & Show an image

```python
img2= cv2.imread('../Images/2.jpg')
cv2.namedWindow('It is Sanji', cv2.WINDOW_NORMAL)
cv2.imshow('It is Sanji', img2)
cv2.waitKey()
cv2.destroyAllWindows()

print(img2.shape)
```

Make grayScale image

```python
gray_img2= cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
cv2.namedWindow('Gray Sanji', cv2.WINDOW_NORMAL)
cv2.imshow('Gray Sanji', gray_img2)
cv2.waitKey()
cv2.destroyAllWindows()
```

Import grayScale image

```python
img2_gray= cv2.imread('../Images/2.jpg', 0)
cv2.namedWindow('It is Sanji', cv2.WINDOW_NORMAL)
cv2.imshow('It is Sanji', img2_gray)
cv2.waitKey()
cv2.destroyAllWindows()

print(img2_gray.shape)
cv2.imwrite('2-1.jpg', img2_gray)
```

Import grayScale image in PLT

```python
plt.imshow(img2_gray, cmap= 'gray')
```

Crop image(Method 1)

```python
height, width= img2.shape[:2]

cropped_img2= img2[0:500, 100:650]

plt.imshow(cropped_img2[..., ::-1])
cv2.imwrite('2-2.jpg', cropped_img2)
```
