```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Split channels

```python
img3= cv2.imread('../Images/3.jpg')
b, g, r= cv2.split(img3)

plt.Figure(figsize= [20, 5])
plt.subplots_adjust(wspace= 0.7)
plt.subplot(141);plt.imshow(r, cmap= 'gray');plt.title('Red');
plt.subplot(142);plt.imshow(g, cmap= 'gray');plt.title('Green');
plt.subplot(143);plt.imshow(b, cmap= 'gray');plt.title('Blue');

plt.subplot(144);plt.imshow(img3[:, :, ::-1]);plt.title('All');
```

```python
img4= cv2.imread('../Images/4.jpg')
b, g, r= cv2.split(img4)

plt.Figure(figsize= [20, 5])
plt.subplots_adjust(wspace= 0.6)
plt.subplot(141);plt.imshow(r, cmap= 'gray');plt.title('Red');
plt.subplot(142);plt.imshow(g, cmap= 'gray');plt.title('Green');
plt.subplot(143);plt.imshow(b, cmap= 'gray');plt.title('Blue');

plt.subplot(144);plt.imshow(img4[:, :, ::-1]);plt.title('All');
```

Merge channels

```python
merged_img4= cv2.merge([b, g, r])

plt.Figure(figsize= [20, 5])
plt.subplots_adjust(wspace= 0.6)
plt.subplot(141);plt.imshow(r, cmap= 'gray');plt.title('Red');
plt.subplot(142);plt.imshow(g, cmap= 'gray');plt.title('Green');
plt.subplot(143);plt.imshow(b, cmap= 'gray');plt.title('Blue');

plt.subplot(144);plt.imshow(merged_img4[:, :, ::-1]);plt.title('Merged Channels');
```

```python
zeros= np.zeros_like(r, 'uint8')

red= cv2.merge([zeros, zeros, r])
green= cv2.merge([zeros, g, zeros])
blue= cv2.merge([b, zeros, zeros])

plt.subplots_adjust(wspace= 0.6)
plt.subplot(141);plt.imshow(red[..., ::-1], cmap= 'gray');plt.title('Red');
plt.subplot(142);plt.imshow(green[..., ::-1], cmap= 'gray');plt.title('Green');
plt.subplot(143);plt.imshow(blue[..., ::-1], cmap= 'gray');plt.title('Blue');
```

Modify channels(Methode 1)

```python
img5= cv2.imread('../Images/5.jpg')
b, g, r= cv2.split(img5)

plt.subplots_adjust(wspace= 0.5)
plt.subplot(141);plt.imshow(r, cmap= 'gray');plt.title('Red');
plt.subplot(142);plt.imshow(g, cmap= 'gray');plt.title('Green');
plt.subplot(143);plt.imshow(b, cmap= 'gray');plt.title('blue');

plt.subplot(144);plt.imshow(img5[:, :, ::-1]);plt.title('Orginal');
```

```python
modified_b= b+20
merged_img5= cv2.merge([modified_b, g, r])

plt.subplots_adjust(wspace= 0.5)
plt.subplot(141);plt.imshow(r, cmap= 'gray');plt.title('Red');
plt.subplot(142);plt.imshow(g, cmap= 'gray');plt.title('Green');
plt.subplot(143);plt.imshow(modified_b, cmap= 'gray');plt.title('Modified blue');

plt.subplot(144);plt.imshow(merged_img5[:, :, ::-1]);plt.title('All');
```

Modify image(Methode 2)

```python
modified_b2= b.astype('int16') + 40
modified_b2= np.minimum(modified_b2, 255)
modified_b2= modified_b2.astype('uint8')

merged_img5_2= cv2.merge([modified_b2, g, r])

plt.Figure(figsize= [20, 8])
plt.subplots_adjust(wspace= 0.5)
plt.subplot(241);plt.imshow(r, cmap= 'gray');plt.title('Red');
plt.subplot(242);plt.imshow(g, cmap= 'gray');plt.title('Green');
plt.subplot(243);plt.imshow(b, cmap= 'gray');plt.title('blue');
plt.subplot(244);plt.imshow(img5[:, :, ::-1]);plt.title('Orginal');

plt.subplot(247);plt.imshow(modified_b2, cmap= 'gray');plt.title('Modified blue');
plt.subplot(248);plt.imshow(merged_img5_2[:, :, ::-1]);plt.title('Modified');

plt.imsave('Modified.png',merged_img5_2[:, :, ::-1])
```

Modify image(Method 3)

```python
modified_b3 = b
np.all(modified_b3 == cv2.add(b, 20))
merged_img5_3= cv2.merge([modified_b3, g, r])

plt.Figure(figsize= [20, 8])
plt.subplots_adjust(wspace= 0.5)
plt.subplot(241);plt.imshow(r, cmap= 'gray');plt.title('Red');
plt.subplot(242);plt.imshow(g, cmap= 'gray');plt.title('Green');
plt.subplot(243);plt.imshow(b, cmap= 'gray');plt.title('blue');
plt.subplot(244);plt.imshow(img5[:, :, ::-1]);plt.title('Orginal');

plt.subplot(247);plt.imshow(modified_b3, cmap= 'gray');plt.title('Modified blue');
plt.subplot(248);plt.imshow(merged_img5_3[:, :, ::-1]);plt.title('Modified');
```
