```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Draw using mouse

```python
drawing = False
color = (0, 255, 0)

def brush(event, x, y, flags, param):
    global drawing, color
    
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
           cv2.circle(img, (x, y), 5, color, -1)
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        
img = np.zeros((512, 512, 3), np.uint8)
cv2.namedWindow('Image draw')
cv2.setMouseCallback('Image draw', brush)

while True:
    cv2.imshow('Image draw', img)
    k = cv2.waitKey(1) & 0xFF
    if k == ord('b'):
        color = (255, 0, 0)
    elif k == ord('r'):
        color =  (0, 0, 255)
    elif k == ord('g'):
        color = (0, 255, 0)
    elif k == 27 & 0xFF:
        break    
    
cv2.destroyAllWindows()
plt.imshow(img[..., ::-1])
```
